### Task 1: Merge the 12 months of sales data into a single csv file

```python

files = []

for file in os.listdir("./dataset/sales-data"):
    files.append(file)

all_months_data = pd.DataFrame()

for file in files:
    path = "./dataset/sales-data/" + file
    df = pd.read_csv(path)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("./dataset/sales-data/all_data.csv")
```
